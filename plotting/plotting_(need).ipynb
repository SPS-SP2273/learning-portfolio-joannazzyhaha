### Things to note

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 9]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [5, 2, 7, 11, 3, 7, 8]
x_error, y_error = .1, 0.75
# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed',
        lw=3)
plt.plot(x, y_2, label='another line',
        color='blue', linestyle='dotted')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.1)
plt.legend(loc='center')
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green", linestyle='dashed')
```

# 2 Better with NumPy

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.1)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 0.05, 0.10, 0.22, 0.34, 0.45, 1.86])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

### Things to note

```python
x = np.linspace(0, 6, 50)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                  # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right') # For y-axis 2

plt.savefig('simple-plot.png', dpi=100)
```

# 4 A real example: Global Warming

```python
with open('Complete_TAVG_daily.txt', 'r') as file:
    file_content = file.read()

print(file_content)
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.style.use('fivethirtyeight')
```
